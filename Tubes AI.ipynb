```python
import random

# Base Stats
baseATK = 335 + 608
bonusATK = 0.496
dmgMultiplier = 4.85
critRate = 0.24
critMultiplier = 0.5
bonusDMG = 0
baseHP = 12981
bonusHP = 0

# Artifact Main Stats
baseATK += 311
bonusATK += 0.466
bonusDMG += 0.466
critRate += 0.311

population = []
currentFitness = []
parent = []
children = []

def initializePopulation():
    for i in range(0,6):
        chromosome = []

        for j in range(0, 10):
            chromosome.append(random.randint(0,1))

        # Repair Kromosom
        while chromosome.count(1) > 5:
            repairIndex = chromosome.index(1)
            chromosome[repairIndex] = 0

        population.append(chromosome)

def defineChromosome():
    currentFitness.clear()
    for i in population:
        subTotalCritMultiplier = 0
        subTotalbonusATK = 0
        subTotalCritRate = 0
        subTotalbonusHP = 0
        if i[0] == 1: subTotalCritMultiplier    += 0.3
        if i[1] == 1: subTotalbonusATK          += 0.2
        if i[2] == 1: subTotalCritRate          += 0.15
        if i[3] == 1: subTotalCritMultiplier    += 0.3
        if i[4] == 1: subTotalbonusATK          += 0.5
        if i[5] == 1: subTotalCritRate          += 0.15
        if i[6] == 1: subTotalCritMultiplier    += 0.45
        if i[7] == 1: subTotalbonusATK          += 0.1
        if i[8] == 1: subTotalCritRate          += 0.3
        if i[9] == 1: subTotalbonusHP           += 0.3

        totalCritMultiplier = critMultiplier + subTotalCritMultiplier
        totalBonusATK = bonusATK + subTotalbonusATK
        totalCritRate = critRate + subTotalCritRate

        if totalCritRate > 1: totalCritRate = 1
        totalBonusHP = bonusHP + subTotalbonusHP

        calculateFitness(totalCritMultiplier, totalBonusATK, totalCritRate)

def calculateFitness(totalCritMultiplier, totalBonusATK, totalCritRate):
    finalATK = baseATK * (1 + totalBonusATK)
    baseDamage = (dmgMultiplier * finalATK) * (1 + bonusDMG)
    critDamage = baseDamage * (1 + totalCritMultiplier)
    averageDamage =  (critDamage * totalCritRate) + (baseDamage * (1 - totalCritRate))
    currentFitness.append(int(averageDamage))

def selection():
    parent.clear()
    for i in range(0, 6):
        randomNumber = random.randint(0, 1000)/1000
        # print(randomNumber)
        if randomNumber < sum(currentFitness[0:1])/sum(currentFitness):
            parent.append(population[0])
        elif randomNumber < sum(currentFitness[0:2])/sum(currentFitness):
            parent.append(population[1])
        elif randomNumber < sum(currentFitness[0:3])/sum(currentFitness):
            parent.append(population[2])
        elif randomNumber < sum(currentFitness[0:4])/sum(currentFitness):
            parent.append(population[3])
        elif randomNumber < sum(currentFitness[0:5])/sum(currentFitness):
            parent.append(population[4])
        elif randomNumber < sum(currentFitness[0:6])/sum(currentFitness):
            parent.append(population[5])

        # print(parent[i])

        # Cek pasangan yang sama
        while (i % 2 == 1) and (parent[i] == parent[i-1]):
            randomNumber = random.randint(0, 1000)/1000
            # print(randomNumber,"retry")
            if randomNumber < sum(currentFitness[0:1])/sum(currentFitness):
                parent[i] = population[0]
            elif randomNumber < sum(currentFitness[0:2])/sum(currentFitness):
                parent[i] = population[1]
            elif randomNumber < sum(currentFitness[0:3])/sum(currentFitness):
                parent[i] = population[2]
            elif randomNumber < sum(currentFitness[0:4])/sum(currentFitness):
                parent[i] = population[3]
            elif randomNumber < sum(currentFitness[0:5])/sum(currentFitness):
                parent[i] = population[4]
            elif randomNumber < sum(currentFitness[0:6])/sum(currentFitness):
                parent[i] = population[5]
            # print(parent[i])
    print("Parent")
    for i in range(0, 6):
        print(parent[i])

def crossover(x):
    children.clear()
    recombinationPoint = 5
    for i in range(0, 3):
        randomNumber = random.randint(0, 1000)/1000
        if randomNumber < x:
            children.append(parent[i*2][:recombinationPoint] + parent[i*2+1][recombinationPoint:])
            children.append(parent[i*2+1][:recombinationPoint] + parent[i*2][recombinationPoint:])
        else:
            children.append(parent[i*2])
            children.append(parent[i*2+1])

    # Repair Kromosom
    for i in children:
        while i.count(1) > 5:
            repairIndexChildren = i.index(1)
            i[repairIndexChildren] = 0

    # for j in range(0, 6):
    #     print("Children", j, "=", children[j])

def mutation(x):
    for i in range(0, 6):
        for j in range(0, 10):
            randomNumber = random.randint(0, 1000)/1000
            if randomNumber < x:
                # print("Sebelum Mutasi")
                # print(children[i])
                if children[i][j] == 0: children[i][j] = 1
                elif children[i][j] == 1: children[i][j] = 0
                # print("Setelah Mutasi!!")
                # print(children[i])

    # Repair Kromosom
    for i in children:
        while i.count(1) > 5:
            repairIndexChildren = i.index(1)
            i[repairIndexChildren] = 0
    print("Children")
    for j in range(0, 6):
      print(children[j])

bestFitness = 0
iteration = int(input("Masukkan Berapa Generasi: "))
silang = float(input("Masukkan probabilitas crossover: "))
mutasi = float(input("Masukkan probabilitas mutasi: "))
print("=============================\n")

initializePopulation()
print("Initial Chromosome")
for i1 in range(0, 6):
    print(population[i1])
for i1 in range(1, iteration):
    defineChromosome()
    # print("hah",currentFitness)
    if max(currentFitness) > bestFitness:
        print("Best Berubah")
        bestFitness = max(currentFitness)
        # bestChromosome = population[currentFitness.index(max(currentFitness))]
        bestChromosomeIndex = currentFitness.index(max(currentFitness))
        bestChromosome = population[bestChromosomeIndex]
    print("Current Fitness:", currentFitness)
    print("Best Fitness =", bestFitness)
    print("Best Chromosome =", bestChromosome)
    print("=============================")
    print(f"Generasi ke-{i1+1}")
    selection()
    crossover(silang)
    mutation(mutasi)
    for i in range(0,6):
      population[i]=children[i]
    # print("cek",population)
    # defineChromosome()
# print("hah2",currentFitness)
defineChromosome()
if max(currentFitness) > bestFitness:
    print("Best Berubah")
    bestFitness = max(currentFitness)
    # bestChromosome = population[currentFitness.index(max(currentFitness))]
    bestChromosomeIndex = currentFitness.index(max(currentFitness))
    bestChromosome = population[bestChromosomeIndex]
# print("Best Chromosome Index", bestChromosomeIndex)
print("Best Fitness =", bestFitness)
print("Best Chromosome =", bestChromosome)
```
